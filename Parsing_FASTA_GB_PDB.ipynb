```python
from Bio import SeqIO #buat FASTA sama GB
from Bio.PDB import PDBParser #kuhus buat PDB
#SMRCA

#Structure
#   |-> Model
#       |-> Chain
#           |-> Residue
#               |-> Atom
```

```python
fasta = SeqIO.parse("./rcsb_pdb_9FNS.fasta", "fasta")
seqA = []
for record in fasta:
    print(record.id)
    print(record.name)
    print(record.seq)
    seqA.append(record.seq)
    print(seqA)
```

```python
gb = SeqIO.parse("./sequence.gb", "gb")
seqA = []
for record in gb:
    print(record)
    print(record.id)
    print(record.name)
    print(record.seq)
    seqA.append(record.seq)
    print(seqA)
```
